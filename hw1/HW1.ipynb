<br>1.2:
<br>
<br>show: $\forall \epsilon \in (0,p] : Pr[|X-p| \gt \epsilon] \lt 2^{-\omega(\epsilon^2 n)}$ 
<br>
<br>will bound: $Pr[X-p \gt \epsilon]$ other side is similar.
<br> $= Pr[\bar{X} \gt \epsilon]$
<br> $= Pr[e^{\lambda \bar{X}} \gt e^{\lambda \epsilon}]$ (for any positive lambda)
<br> $= Pr[e^{\lambda \bar{X}} \gt \frac{E[e^{\lambda \bar{X}}]}{E[e^{\lambda \bar{X}}]}e^{\lambda \epsilon}]$
<br> $ \lt \frac{E[e^{\lambda \bar{X}}]}{e^{\lambda \epsilon}}$ (from Markov's inequality choosing as $\alpha$: $\frac{e^{\lambda \epsilon}}{E[e^{\lambda \bar{X}}]}$)
<br> $ \le \frac{e^{\frac{\lambda^2}{n}p(1-p)}}{e^{\lambda \epsilon}}$ (from lemma $(*)$ which we show later)
<br> $ = e^{-\frac{\epsilon^2}{4p(1-p)}n}$ (setting lambda as $\frac{n\epsilon}{2p(1-p)}$)
<br> $ = 2^{-\Omega(\epsilon^2 n)}$
<br>
<br>$(*)$ lemma: $E[e^{\lambda \bar{X}}] \le e^{\frac{\lambda^2}{n}p(1-p)}$ :
<br>
<br>$E[e^{\lambda \bar{X}}]$
<br>$= E[e^{\frac{\lambda}{n} \sum^n_i \bar{X_i}}]$
<br>$= \Pi^n_i E[ e^{\frac{\lambda}{n} \bar{X_i}}]$ (from the fact that $\bar{X_i}$ are independent)
<br>$= \Pi^n_i (p \cdot e^{\frac{\lambda}{n} (1-p)} + (1-p) \cdot e^{\frac{\lambda}{n} (-p)})$
<br>$\le (p \cdot (1 + \frac{\lambda}{n}(1-p) + \frac{\lambda^2}{n^2}(1-p)^2) + (1-p) \cdot (1 + \frac{\lambda}{n}p + \frac{\lambda^2}{n^2}p^2))^n$
<br>$= (1 + \frac{\lambda^2}{n^2}p(1-p))^n$
<br>$= (1 + \frac{\frac{\lambda^2}{n}p(1-p)}{n})^n \rightarrow e^{\frac{\lambda^2}{n}p(1-p)}$
<br> so for high enough $n$ we get what we need.

<br> the interaction is repeated sequentially t times and V decides by majority.

<br>completeness: 
<br>let $x \in L$ 
<br>$Pr[(P^t,V^t)(x) = 1]$
<br>$= Pr[V \text{ accepts } \frac{t}{2} \text{ of runs or more}]$
<br>$= Pr[\frac{1}{t} \sum^t_i E_i \ge \frac{1}{2}$] (where $E_i$ denotes indicator for the event that V accepted in the i'th round)
<br>$ \ge Pr[|\frac{1}{t} \sum^t_i E_i - \frac{2}{3}| \le \frac{1}{6}]$
<br>$ = 1 - Pr[|\frac{1}{t} \sum^t_i E_i - \frac{2}{3}| \gt \frac{1}{6}]$ 
<br>$ \gt 1 - e^{-\frac{\frac{1}{6}^2}{4\frac{2}{3}(1-\frac{2}{3})}t}$ (from Chernoff proof in 1.2)
<br>$ \gt 1 - e^{-\frac{t}{32}}$
<br>$ = 1 - 2^{-\Omega(t)}$
<br>
<br>soundness:
<br>let $x \notin L$
<br>$Pr[(P^t,V^t)(x) = 1]$
<br>$= Pr[\frac{1}{t} \sum^t_i E'_i \ge \frac{1}{2}]$ ($E'_i$ : V accepts in the i'th run now that $x \notin L$)
<br>$ \le Pr[|\frac{1}{t} \sum^t_i E'_i - \frac{1}{3}| \gt \frac{1}{6}]$ ($A \rightarrow B \Rightarrow Pr(A) \le Pr(B)$)
<br>$ \le 1 - e^{-\frac{\frac{1}{6}^2}{4\frac{2}{3}(1-\frac{2}{3})}t}$       (Chernoff)
<br>$ = 2^{-\Omega(t)}$

<br>2.
<br>
<br>show $IP^{ps} = NP$
<br>
<br>$(I)$ $\; NP \subseteq IP^{ps}$
<br>Let $L \in NP$
<br>exists polytime $V$ and polynomial $l$ such that
<br>$\forall x \in L \; \exists m \in \{0,1\}^{l(|x|)} : V(x,m) = 1$
<br>$\forall x \notin L \; \forall m^* \in \{0,1\}^{l(|x|)} : V(x,m*) = 0$
<br>
<br>protocol: 
<br>$(P_1,V_1)$ where given $x$, $P_1$ searches for $m \in \{0,1\}^{l(|x|)}$ such that $V(x,m) = 1$
<br>if $P_1$ finds such $m$ it sends is to $V_1$ which accepts if and only if $V(x,m) = 1$.
<br>
<br>completeness:
<br>let $x \in L$ 
<br>$Pr[(P_1,V_1)(x) = 1]$
<br>$= Pr[P_1 \text{ finds } m \text{ such that } V(x,m) = 1] = 1$ ($P_1$ is unbounded so finds such $m$ iff exists)
<br>$ \ge \frac{2}{3}$
<br>
<br>perfect soundness:
<br>let $x \notin L$ 
<br>$Pr[(P_1,V_1)(x) = 1]$
<br>$= Pr[P_1 \text{ finds } m \text{ such that } V(x,m) = 1] = 0$ (such $m$ does not exist)
<br>
<br>
<br>$(II)$ $\; IP^{ps} \subseteq NP$
<br>Let $L \in IP^{ps}$
<br>exists protocol $(P,V)$ ($V$ PPT, $P$ unbounded) such that:
<br>if $x \in L : Pr[(P,V)(x) = 1] \ge \frac{2}{3}$
<br>if $x \notin L \; \forall P^*: Pr[(P^*,V)(x) = 1] = 0$
<br>
<br>can think of $V$ as deterministic poly machine that takes randomness string $\alpha$ as input. 
<br>
<br>let $x \in L$ , $Pr[(P,V)(x) = 1] \ge \frac{2}{3}$ 
<br>$\Rightarrow \; \exists \alpha$ such that $(P,V_{\alpha})(x) = 1$
<br>the interaction is polynomial in $|x|$ ($V_{\alpha}$ is poly time)
<br>and so we denote the interaction as $m$, then $V_{\alpha}(x,m) = 1$ (point 1)
<br>
<br>let $x \notin L$, with the same $\alpha$ as before $\forall P^*$
<br>$(P,V_{\alpha})(x) = 0$ (because $\forall P^* \; Pr_{\alpha}[(P^*,V)(x) = 1] = 0$)
<br>so no $m$ of polynomial length could be sent to $V_{\alpha}$ to make it accept on $x$
<br>$\Rightarrow \; \forall m^* \; V_{\alpha}(x,m^*) = 0$ (point 2)
<br>
<br>from point 1 and 2 we conclude $L \in NP$

<br>3.1
<br>Find $k$ such that $V = V^k_{\text{naive}}$ satisfies:
<br>if $|S| \ge t: \;\;\; Pr[V()=1] \ge 0.99$
<br>if $|S| \le \frac{t}{100}: \;\;\; Pr[V()=1] \le 0.01$
<br>
<br>Precalculations:
<br>$Pr[V()=1] = Pr[|\{i|u_i \in S\}| \ge \frac{kt}{2|U|}]$
<br>(denote $E_i$ the indicator that $u_i \in S$)
<br>$= Pr[\sum_i^k E_i \ge \frac{kt}{2|U|}]$
<br>$= \frac{1}{k} Pr[\sum_i^k E_i \ge \frac{t}{2|U|}]$
<br>$= Pr[\bar{E} \ge \frac{t - 2|S|}{2|U|}]$ (using the notation convention from question 1)
<br>
<br>Soundness:
<br>$Pr[V()=1] = Pr[\bar{E} \ge \frac{t - 2|S|}{2|U|}]$
<br>$\le Pr[\bar{E} \ge \frac{0.98t}{2|U|}]$     (using $|S| \le \frac{t}{100}$)
<br>$\le e^{-\frac{(0.49\frac{t}{|U|})^2}{4\frac{|S|}{|U|}(1-\frac{|S|}{|U|})}k} = 0.01$
<br>$\Rightarrow \: k = -log(0.01) \cdot \frac{4}{(0.49)^2} \cdot \frac{|S|(|U|-|S|)}{t^2}$
<br>
<br>Completeness:
<br>$Pr[V()=1] = Pr[\bar{E} \ge \frac{t - 2|S|}{2|U|}]$
<br>$\ge Pr[\bar{E} \ge \frac{-t}{2|U|}]$     (using $|S| \ge t$)
<br>$= 1 - Pr[\bar{E} \lt \frac{-t}{2|U|}]$
<br>$= 1 - Pr[|\bar{E}| \gt \frac{t}{2|U|}]$
<br>$\ge 1 - e^{-\frac{(\frac{t}{2|U|})^2}{4\frac{|S|}{|U|}(1-\frac{|S|}{|U|})}k} = 0.99$
<br>$\Rightarrow e^{-\frac{t^2k}{16|S|(|U|-|S|)}} = 0.01$
<br>$\Rightarrow \: k = -16 \cdot log(0.01) \cdot \frac{|S|(|U|-|S|)}{t^2}$
<br>
<br>
<br> So to acheive both completeness and soundness we take the higher $k$:
<br>$k = -log(0.01) \cdot \frac{4}{(0.49)^2} \cdot \frac{|S|(|U|-|S|)}{t^2}$

<br>describe PPT verifier $V(A,B,C)$:
<br>sample $r \leftarrow \mathbb{F}^n$ ($O(n)$)
<br>$r_B = B \cdot r$ $\;\;\;\;\;\;\;\;$ ($O(n^2)$)
<br>$r_{AB} = A \cdot r_B$ $\;\;\;\;\;\;$ ($O(n^2)$)
<br>$r_C = C \cdot r$ $\;\;\;\;\;\;\;\;$($O(n^2)$)
<br>if $r_{AB} = r_C$ accept $\;\;\;\;\;\;\;\;$ ($O(n)$)
<br>otherwise reject
<br>
<br>
<br>$V$ runs in $O(n^2)$
<br>
<br>completeness:
<br>$AB = C \Rightarrow A(Br) = Cr \Rightarrow V(A,B,C) = 1 \Rightarrow Pr[V(A,B,C) = 1]=1$
<br>
<br>
<br>soundness:
<br>$ AB \neq C \Rightarrow AB - C \neq 0$
<br>$ Pr[V(A,B,C) = 1] = Pr[ABr = Cr] = Pr[(AB - C)r = 0]$
<br>$ = Pr[\forall i \in [n] : \langle row_i(AB-C),r \rangle = 0]$
<br>( $AB-C \neq 0$ so there exists some $k \in [n]$ for which $row_k(AB-C) \neq \vec{0}$ )
<br>$\le Pr[\langle row_k(AB-C),r \rangle = 0]$
<br>$= \frac{1}{|\mathbb{F}|}$ (from 4.2)
<br>$\le \frac{1}{2}$ (a field consists of at least 2 distinct elements)
