$P: \mathbb{F}^n \rightarrow \mathbb{F}$\
$n$ variate, $d$ total degree.\
\
$Pr_{z \in \mathbb{F}^n}[P(z) = 0] \leq \frac{d}{|\mathbb{F}|}$

1. induction on $n$:\
\
base case: $n=1$ $P: \mathbb{F} \rightarrow \mathbb{F}$, $P \not\equiv 0$ univariate polynomial over field $\mathbb{F}$ having degree $d$.\
(which is also its total degree). Thus from the Fundamental Theorem of Algebra $P(z) = 0$ has no more than $d$ solutions, i.e.\
$Pr_{z \in \mathbb{F}^n}[P(z) = 0] \leq \frac{d}{|\mathbb{F}|}$\
\
step ($n+1$): assume for any $P': \mathbb{F}^n \rightarrow \mathbb{F}$ of total degree $d$:\
 $Pr_{z \in \mathbb{F}^n}[P'(z) = 0] \leq \frac{d}{|\mathbb{F}|}$\
let $P: \mathbb{F}^{n+1} \rightarrow \mathbb{F}$ be polynomial of total degree $d$.\
$Pr_{z \in \mathbb{F}^{n+1}}[P'(z_1, ... , z_{n+1}) = 0]$\
$\leq \sum_{a \in \mathbb{F}} Pr_{z \in \mathbb{F}^{n+1}}[P'(z_1, ... , z_{n+1}) = 0 \;|\; z_1 = a] \cdot Pr_{z_1 \in \mathbb{F}}[z_1 = a]$\
$= \sum_{a \in \mathbb{F}} Pr_{z \in \mathbb{F}^n}[P'(a,z_1, ... , z_n) = 0] \cdot Pr_{z \in \mathbb{F}}[z = a]$\
$\leq \sum_{a \in \mathbb{F}} \frac{d}{|\mathbb{F}|} \cdot \frac{1}{|\mathbb{F}|} = \frac{|\mathbb{F}|}{|\mathbb{F}|} \cdot \frac{d}{|\mathbb{F}|}  = \frac{d}{|\mathbb{F}|}$

2. let $n \in \mathbb{N}$, $d \leq |\mathbb{F}| - 1$\
\
claim:\
$p(z_1, ... , z_n) = \Pi_{k=1}^d(z_1 - k)$\
satisfies $Pr_{z \in \mathbb{F}^n}[p(z) = 0] = \frac{d}{|\mathbb{F}|}$\
\
Thats because $p(z) = 0$ iff $1\leq z_1 \leq d$.\
and randomly chosen $z$ has $\frac{d}{|\mathbb{F}|}$ probability of its first coordinate to satisfy $1\leq z_1 \leq d$.

$A \in_R \{0,1\}^{n \times k}$\
$C : \{0,1\}^k \rightarrow \{0,1\}^n$\
$C(x) = Ax$\
\
Let $\varepsilon \gt 0$, assume $n = \Omega(\frac{k}{\varepsilon^2})$\
show $C$ has relative distance of at least $\frac{1}{2} - \varepsilon$ with probability $0.99$.\
\
$C$ is linear so enough to show for $x=0\;\;$: $\;\;Pr_A[\forall x \neq 0 \;:\; \bar\omega(C(x)) \geq \frac{1}{2}-\varepsilon] \geq 0.99$\
equivalently: $\;\;Pr_A[\exists x \neq 0 \;:\; \bar\omega(C(x)) \lt \frac{1}{2}-\varepsilon] \lt 0.01$\
\
analyze for for some fixed $x \neq 0$:\
$Pr_A[\bar\omega(C(x)) \lt \frac{1}{2}-\varepsilon]$\
$ = Pr_A[\bar\omega(Ax) \lt \frac{1}{2}-\varepsilon]$\
$ = Pr_A[\sum_i^n (Ax)_i \lt n(\frac{1}{2}-\varepsilon)]$\
\

notation $E_i$ is an indicator for the event $(Ax)_i = 1$.\

$ = Pr[\frac{\sum E_i}{n} \lt \frac{1}{2} - \varepsilon]$\
