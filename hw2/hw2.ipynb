<br>$P: \mathbb{F}^n \rightarrow \mathbb{F}$ 
<br>$n$ variate, $d$ total degree.
<br>
<br>$Pr_{z \in \mathbb{F}^n}[P(z) = 0] \leq \frac{d}{|\mathbb{F}|}$
<br>
<br>1. induction on $n$:
<br>
<br>base case: $n=1$ $P: \mathbb{F} \rightarrow \mathbb{F}$, $P \not\equiv 0$ univariate polynomial over field $\mathbb{F}$ having degree $d$. 
<br>(which is also its total degree). Thus from the Fundamental Theorem of Algebra $P(z) = 0$ has no more than $d$ solutions, i.e. 
$Pr_{z \in \mathbb{F}^n}[P(z) = 0] \leq \frac{d}{|\mathbb{F}|}$
<br>
<br>step ($n+1$): assume for any $P': \mathbb{F}^n \rightarrow \mathbb{F}$ of total degree $d$:
 $Pr_{z \in \mathbb{F}^n}[P'(z) = 0] \leq \frac{d}{|\mathbb{F}|}$
<br>let $P: \mathbb{F}^{n+1} \rightarrow \mathbb{F}$ be polynomial of total degree $d$.
<br>$Pr_{z \in \mathbb{F}^{n+1}}[P'(z_1, ... , z_{n+1}) = 0]$
<br>$\leq \sum_{a \in \mathbb{F}} Pr_{z \in \mathbb{F}^{n+1}}[P'(z_1, ... , z_{n+1}) = 0 \;|\; z_1 = a] \cdot Pr_{z_1 \in \mathbb{F}}[z_1 = a]$
<br>$= \sum_{a \in \mathbb{F}} Pr_{z \in \mathbb{F}^n}[P'(a,z_1, ... , z_n) = 0] \cdot Pr_{z \in \mathbb{F}}[z = a]$
<br>$\leq \sum_{a \in \mathbb{F}} \frac{d}{|\mathbb{F}|} \cdot \frac{1}{|\mathbb{F}|} = \frac{|\mathbb{F}|}{|\mathbb{F}|} \cdot \frac{d}{|\mathbb{F}|}  = \frac{d}{|\mathbb{F}|}$
<br>
<br>
<br>2. let $n \in \mathbb{N}$, $d \leq |\mathbb{F}| - 1$
<br>
<br>claim: 
<br>$p(z_1, ... , z_n) = \Pi_{k=1}^d(z_1 - k)$
<br>satisfies $Pr_{z \in \mathbb{F}^n}[p(z) = 0] = \frac{d}{|\mathbb{F}|}$
<br>
<br>Thats because $p(z) = 0$ iff $1\leq z_1 \leq d$.
<br>and randomly chosen $z$ has $\frac{d}{|\mathbb{F}|}$ probability of its first coordinate to satisfy $1\leq z_1 \leq d$.
<br>

1.2 .
=======
lemma 1.2:
<br>guildlines:
<br>
<br>guildline 1. define $I: \mathbb{F}^{2n} \rightarrow \mathbb{F}^n$ so that $I$ is a multilinear individual degree $1$ polynomial s.t:
<br>$\forall x, x' \in \{0,1\}^n$
<br>if $x=x'$ then $I(x,x') = 1$ else $I(x,x') = 0$ 
<br>
<br>claim $I(z) = \Pi_i^n (1 - z_i - z_{i+n} + 2z_iz_{i+n})$ satisfies the above.
<br>$I$ is indeed multilinear.
<br>
<br>if $\exists i \in [n]: z_i \neq z_{i+n}$ then $1 - z_i - z_{i+n} + 2z_iz_{i+n} = 1-1 = 0$ so $I(z) = 0$
<br>
<br>else $\forall i \in [n]:$ either $ z_i = z_{i+n} = 1$ and then $1 - z_i - z_{i+n} + 2z_iz_{i+n} = 1-2+2 = 1$ so $I(z) = 1$
<br>or $ z_i = z_{i+n} = 0$ and then $1 - z_i - z_{i+n} + 2z_iz_{i+n} = 1-0 = 1$ so $I(z) = 1$
<br>
<br>guildline 2. observe that if $\forall x \in \{0,1\}^n$ : $Q(x) = 0$ then $\forall z \in \mathbb{F}^n$ it holds that 
$\sum_{x \in \{0,1\}^n} Q_z(x) = 0$ where $Q_z(x) = Q(x) \cdot I(x,z)$
<br>
<br>guildline 3. $Pr_{z}[\sum_{x \in \{0,1\}^n} Q_z(x) = 0 \;|\; \exists x \in \{0,1\}^n : Q(x) \neq 0]$
<br>$ = Pr_{z}[\sum_{x \in \{0,1\}^n} Q(x) \cdot I(x,z) = 0 \;|\; \exists x \in \{0,1\}^n : Q(x) \neq 0]$
<br>$\leq \frac{nd + 1}{|\mathbb{F}|}$ from polynomial identity lemma and $(*)$,$(**)$ of which proof is later 
<br>
<br>$(*)$ $Q$ individual degree $d$, $I$ individual degree 1.
<br>$t_z(x) = \sum_{x \in \{0,1\}^n} Q(x) \cdot I(x,z)$ is total degree of at most $nd +1$
<br>
<br>$(**)$ $\exists x' \in \{0,1\}^n : Q(x) \neq 0$ so for $z = x'$ : $\sum_{x \in \{0,1\}^n} Q(x) \cdot I(x,x') = Q(x') \cdot 1 \neq 0$
<br>so not the $0$ polynomial
<br>
<br>conclude:
<br>$Pr_{z}[\sum_{x \in \{0,1\}^n} Q_z(x) \neq 0 \;|\; \exists x \in \{0,1\}^n : Q(x) \neq 0] \geq 1 - \frac{nd + 1}{|\mathbb{F}|}$
<br>
<br>
<br>guildline 4. given $Q: \mathbb{F}^n \rightarrow \mathbb{F}$, $V$ chooses $z \leftarrow \{0,1\}^n$ randomly and sends ot to $P$.
<br>
<br>initiate sumcheck $V_{sc},P_{sc}$ on $Q_z(x) = Q(x) \cdot I(x,z)$
<br>(given orcale access to $Q$, $V$ supplies $V_{sc} with orcale access to $Q_z$ by accesssing $Q(x)$ and multiplying with $I(x,z)$)
<br>$V$ accepts or rejects accordingly.
<br>
<br>
<br>completeness: 
<br>assume $Q(x) = 0 \;\;\; \forall x \in \{0,1\}^n$
<br>
<br>$Q_z(x) = Q(x) \cdot I(x,z) = 0 \;\;\; \forall x \in \{0,1\}^n$
<br>
<br>$Pr[V \text{ accepts}] = Pr[V_{sc} \text{ accepts}] = 1$
<br>
<br>
<br>soundness: 
<br>assume $\exists x \in \{0,1\}^n : Q(x) \neq 0$
<br>$Pr[V \text{ accepts}] = Pr[V_{sc} \text{ accepts}]$
<br>$= Pr[V_{sc} \text{ accepts} \;|\; \sum Q_z(x) = 0] \cdot Pr[\sum Q_z(x) = 0] + Pr[V_{sc} \text{ accepts} \;|\; \sum Q_z(x) \neq 0] \cdot Pr[\sum Q_z(x) \neq 0]$
<br>$\leq 1 \cdot \frac{nd + 1}{|\mathbb{F}|} + \frac{nd + 1}{|\mathbb{F}|}\cdot 1 = O(\frac{nd}{|\mathbb{F}|})$ 
<br>
<br>
<br>complexity:
<br>$V$ can compute $I$ in $poly(n,d,log|\mathbb{F}|)$ so that will be the cost of every orcale query by $V_{sc}$ (considering also $V$ queries its orcale in $O(1)$).
<br>$V_{sc}$'s communication is $poly(n)$, so overall the protocol runs in $poly(n,d,log|\mathbb{F}|)$.
<br>
<br>
<br>#queries:
<br>$V$ queries $Q$ exactly once for every time $V_{sc}$ queries $Q_z$ which is $n$ times (number of coordinates of input to $Q_z).
<br>So $V$ queries $Q$ $\;n$ times.

<br>$C : \{0,1\}^k \rightarrow \{0,1\}^n$
<br>$C(x) = Ax$
<br>
<br>Let $\varepsilon \gt 0$, assume $n = \Omega(\frac{k}{\varepsilon^2})$
<br>show $C$ has relative distance of at least $\frac{1}{2} - \varepsilon$ with probability $0.99$.
<br>
<br>
<br>$C$ is linear so enough to show for $x=0$:
<br>$Pr_A[\bar\omega(C(x)) \geq \frac{1}{2}-\varepsilon] \geq 0.99$

<br>
<br>$Pr_A[\bar\omega(C(x)) \geq \frac{1}{2}-\varepsilon]$
<br>$= Pr_A[\bar\omega(Ax) \geq \frac{1}{2}-\varepsilon]$

<br>notation $E_i$ is an indicator for the event (Ax)_i = 1.
